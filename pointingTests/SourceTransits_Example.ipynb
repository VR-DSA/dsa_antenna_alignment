```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.time import Time
from astropy import units as u
```

```python
# load data in 
data = np.load('20191023_2302_data_B.npz', allow_pickle=True)
```

```python
print(data.files)
```

```python
# set up variables from file

# powers
spectra = data["data"]
print(spectra.shape)

# time stamps
times_mjd = data['mjd'] # start time in MJD
tsamp = data['tsamp'] # sampling time in seconds
print(times_mjd)
print(tsamp)


# frequency channels
fsamp = data["fsamp"]/(1e6) # convert fsamp from Hz to MHz
f0 = data['f0']/(1e6) # convert f0 from Hz to MHz
fchanns = np.array(np.arange(spectra.shape[1]))
fchanns = f0 - fchanns*fsamp
```

```python
# plot spectra for one timestamp
plt.figure(0)
plt.plot(fchanns[:], spectra[0][:])
plt.ylabel('Power');
plt.xlabel('Freq (MHz)');
```

```python
# use RFI free channels as described in memo
intf = np.argwhere((fchanns>1403) & (fchanns<1461.6))
intf = np.ravel(intf)

# integrate power across this slice of frequency channels
spectra_ = spectra[:,intf]
spec_int = np.mean(spectra_,1)

# set up times as seconds since observation started
times = np.array(np.arange(spectra.shape[0]))*tsamp

# re-bin samples by average sets of 32 samples
mean_times = np.mean(times.reshape(-1, 32), axis=1)
mean_spec = np.mean(spec_int.reshape(-1, 32), axis=1)
```

```python
plt.figure(0)
plt.plot(mean_times, mean_spec)
plt.ylabel('Power (raw values)', fontsize=13);
plt.xlabel('Observation Elapsed Time (s)', fontsize=13);
```

```python
tstart_mjd = Time(times_mjd, format='mjd')
t_max = tstart_mjd + (mean_times[np.argmax(mean_spec)]*u.second)

print(tstart_mjd)
print(t_max)
transit_expected = Time(58779.98036774434, format='mjd')

tdiff = t_max - transit_expected
print(tdiff)
print(tdiff.to(u.second))
```
